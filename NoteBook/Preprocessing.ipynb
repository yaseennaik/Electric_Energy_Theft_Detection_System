```python
# importing the libs Necessary for data preprocessing
import pandas as pd
```

```python
def br():
    print('*'*90)

df=pd.read_csv('../Dataset/Raw_Data.csv')
print("The head of dataset:")
df['Anomaly_Label'] = df['Anomaly_Label'].map({'Normal': 0, 'Abnormal': 1})
print(df.head())
br()


# checking the shape and basic info about the data
print("The shape of the dataset is:")
print(df.shape)
br()
print("The info of the dataset is:")
df.info()
br()

# Removing the Average Past Consumption Column
print("The info of the dataset after removing the 'Avg_Past_Consumption' column:")
df=df.drop('Avg_Past_Consumption',axis=1)
df.info()
br()

# Handling Missing Values
print("Missing Values:\n", df.isnull().sum())
br()

# Checking the duplicates in the dataset
print("The Duplicates are: ")
print(df.duplicated().sum())
br()

# Converting Timestamp to datetime and sorting on the basis of time stamp
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df = df.sort_values(by='Timestamp')

# Adding new column: Avg_of_past12 : Average of pass 12 enteries of the energy consumption
print("The info of the dataset after adding the 'Avg_of_past12' column:")
df['Avg_of_past12'] = df['Electricity_Consumed'].rolling(window=12, min_periods=1).mean()
df.info()
br()

#Saving the processed Dataset
df.to_csv("../Dataset/Processed_data.csv", index=False)

```
