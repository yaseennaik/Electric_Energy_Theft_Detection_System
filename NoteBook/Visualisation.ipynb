```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the data from the csv file
df = pd.read_csv('../Dataset/Processed_data.csv')

# histogram showing the distribution of the features in the dataset
for col in df.select_dtypes(include='number'):
    sns.histplot(data=df,x=col)
    plt.title(col)
    plt.show()
```

```python
for col in ['Electricity_Consumed', 'Temperature', 'Humidity', 'Wind_Speed', 'Avg_of_past12']:
    sns.boxplot(data=df, x='Anomaly_Label', y=col)
    plt.title(f'{col} vs Anomaly Class')
    plt.xlabel('Anomaly Class (0 = Normal, 1 = Anomaly)')
    plt.ylabel(col)
    plt.show()
```

```python
plt.figure(figsize=(5,5))
hmap=df.select_dtypes(include='number').corr()
sns.heatmap(data=hmap,annot=True)
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Timestamp'], df['Electricity_Consumed'], label='Electricity Consumption', color='blue')
plt.scatter(df[df['Anomaly_Label']==1]['Timestamp'],
            df[df['Anomaly_Label']==1]['Electricity_Consumed'],
            color='red', label='Anomalies', marker='x')
plt.title("Electricity Consumption Over Time (Anomalies Highlighted)")
plt.xlabel("Timestamp")
plt.ylabel("Electricity Consumed (kWh)")
plt.legend()
plt.show()
```
